##Embarked

```python
df[df.Embarked.isnull()]
```

```python
#tampilkan proporsi data dari data column Embarked
#ternyata data column EMbarked berupa data chategoric

df.Embarked.value_counts()
```

Bila kita akan melakukn imputation pada column Embarked
maka kita periksa tipe data column Embarked terlebih dahulu
Data pada column Embarked berupa data categoric, maka imputation nya menggunakan modus.
Dari proporsi column Embarked, S adalah data yang sering muncul, maka S merupakan modusnya,

```python
val = df.Embarked.mode().values[0]
df['Embarked'] = df.Embarked.fillna(val)
```

```python
df.Embarked.value_counts()
```

```python
#column Embarked masih berupa type data Object
#untuk memudahkan proses analisis maka kita akan convert type object menjadi type numeric

df.Embarked = df.Embarked.map({'S':0, 'C':1, 'Q':2})
```

```python
df.info()
```

##Age

Terdapat 891 data, namun column Age memiliki missing value karena hanya terdapat 714 data yang tersedia dari yang seharusnya 891

Tampilkan visualisasi kolom nya

```python
df['Age'].plot(kind='hist');
```

```python
sns.distplot(df['Age'])
```

```python
df['Age'].value_counts()
```

Karena kolom Age berdistribusi skewness
maka kita akan melakukan imputation (memasukkan data) pada kolom Age dengan median

```python
val = df.Age.median()
df['Age'] = df.Age.fillna(val)
```

```python
df.info()
```

```python
#untuk melihat visualisasi outlier pada column Age

sns.boxplot(df['Age'])
```

##Cabin

```python
df['Cabin'].value_counts()
```

Terlihat bahwa value column Cabin terlalu banyak data unique nya
dan juga info pada column Cabin tidak terlalu informatif untuk mengetahui data Survived.
maka kita dapat men-drop (menghapus) column Cabin

Dan jika data missing terlalu banyak, lebih banyak daripada data yang terisi, maka data lebih baik di drop, agar hasil data yang teranalisis tidak bias. Namun, harus melalui diskusi terlebih dahulu dengan user.

```python
df.drop('Cabin', axis=1, inplace = True) #untuk menghapus column Cabin
#inplace = True untuk menghapus column Cabin secara permanen, jika tidak ada inpace = True, column Cabin hanya terhapus sementara
```

```python
df.info()
```
